```python
import numpy as np, pandas as pd
from pyteomics import mzxml, pepxml
from statistics import median 

# Input files
mFile = "/Research/Tmp/HL_human.mzXML"
pFile = "/Research/Tmp/HL_human.pep.xml.output"

# Loading pepXML
# Make a relationship between peptides and PSMs (peptide to spectra)
pep2psm = {}
pReader = pepxml.read(pFile)
with pepxml.read(pFile) as pReader:
    for spec in pReader:
        if "search_hit" in spec:
            if spec["start_scan"] != spec["end_scan"]:
                print("Start-scan and end-scan are different")
                break
            scanNum = spec["start_scan"]
            peptide = spec["search_hit"][0]["peptide"]    # Peptide from the 1st hit rank
            if peptide in pep2psm:
                pep2psm[peptide].append(scanNum)
            else:
                pep2psm[peptide] = [scanNum]
                
# df = pepxml.DataFrame(pFile)    # Which is faster?
```

```python
def calculateSnr(arr):
    sortedArr = sorted(arr, reverse=True)    # Descending order
    snr = (sortedArr[1] + sortedArr[5]) / 2    # Average of 2nd and 6th highest signals
    snr = snr / median(sortedArr)    # Divide by the median intensity (noise)
    return snr


def rankReplicates(reps, reader):
    # reps = array of scan numbers (i.e., replicates)
    # reader = mzXML reader
    snr = []
    for rep in reps:
        snr.append(calculateSnr(reader[str(rep)]["intensity array"]))
    reps = [x for _, x in sorted(zip(snr, reps), reverse=True)]
    return reps
```
